```python
model = SentenceTransformer('all-minilm-l6-v2')
```

```python
from sentence_transformers import SentenceTransformer, util
import numpy as np


model = SentenceTransformer('all-minilm-l6-v2')


sentences = [
    "I love machine learning.",
    "The dog jumped over the fence.",
    "Artificial intelligence is fascinating.",
    "Can you pass me the salt, please?",
    "The sun is shining brightly today.",
    "Deep learning models require large datasets.",
    "We need to pick up groceries after work.",
    "Natural language processing is evolving rapidly.",
    "She enjoys reading mystery novels in her free time.",
    "Data science is transforming industries.",
    "The coffee shop down the street serves the best latte.",
    "I plan to visit Japan next summer.",
    "Reinforcement learning can be applied in gaming.",
    "Its raining outside, better take an umbrella.",
    "Neural networks are inspired by the human brain.",
    "My favorite hobby is photography.",
    "Machine learning models are used for image classification.",
    "Shes taking a course on artificial intelligence this semester.",
    "Supervised learning involves labeled data.",
    "I havent seen that movie yet, is it good?"
]


vec = model.encode(sentences)


print("Sentence embeddings:")
for i, sentence in enumerate(sentences):
    print(f"Sentence {i+1}: {vec[i]}")
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reduce dimensions
pca = PCA(n_components=2
          )
reduced = pca.fit_transform(vec)

# Plot
plt.figure(figsize=(8,6))
plt.scatter(reduced[:, 0], reduced[:, 1])

for i, text in enumerate(sentences):
    plt.annotate(text, (reduced[i, 0], reduced[i, 1]))

plt.title("Sentence Embeddings (PCA-reduced)")
plt.show()
```
